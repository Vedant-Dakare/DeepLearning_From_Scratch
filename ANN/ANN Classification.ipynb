```python
import pandas as pd
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
## preprocess the data
## drop the irrelevant columns
df = df.drop(columns=['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df.head()
```

```python
## Encode categorical features
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])
df
```

```python
df['Geography'].unique()
```

```python
## one hot encode 'Geography'
from sklearn.preprocessing import OneHotEncoder
oh_encoder_geo = OneHotEncoder()
geo_encoder = oh_encoder_geo.fit_transform(df[['Geography']])
geo_encoder
```

```python
geo_encoder.toarray()
```

```python
oh_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder.toarray(),columns=oh_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
## Combine onehotencoder columns with the original data
df = pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
df.head()
```

```python
## save the encoder and scalar
with open('label_encoder_gender.okl','wb') as file:
  pickle.dump(label_encoder_gender,file)

with open('onehot_encoder_geo.pkl','wb') as file:
  pickle.dump(oh_encoder_geo,file)
```

```python
## train test split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

## scaele the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```
