```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    'the glass of milk',
    'the cup of tea',
    'you are a good developer',
    'dwag love the high songs',
    'this all is heart settling'
]
```

```python
sent
```

```python
## Definng the vobalulary size
voc_size = 10000
```

```python
## one hot representation 
one_hot_rep = [one_hot(words,voc_size)for words in sent]    ## converting the OHE representation
one_hot_rep
```

```python
## word embedding representation
from tensorflow.keras.layers import Embedding      ## converts words into dense vector representations
from tensorflow.keras.utils import pad_sequences   ## to make the length of sentence same
from tensorflow.keras.models import Sequential     ## to create the model
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## feature reprenstation 
dim=10      ## dimension of feature vector
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))     ## creating the embedding layer
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
